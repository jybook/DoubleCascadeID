```python
import pandas as pd
import os, sys, glob

MC_DIR = '/n/holylfs05/LABS/arguelles_delgado_lab/Lab/IceCube_MC/HNL/parquet/'
SIGNAL_DIRS = ['190301', '190302', '190303', '190304', '190305', '190306', '190307', '190308']
BACKGROUND_DIRS = ['oscNext']
#Both of these have sub-folders 'SRTInIcePulses' and 'truth'

parquet_files = []
for subdir in SIGNAL_DIRS:
    parquet_files.append(sorted(glob.glob(os.path.join(f"{MC_DIR}{subdir}/truth/*.parquet"))))

# Read and concatenate all Parquet files
dfs = [pd.read_parquet(file) for file in parquet_files]
background_df = pd.read_parquet(glob.glob(os.path.join(f"{MC_DIR}{BACKGROUND_DIRS}/truth/*.parquet")))

merged_df = pd.concat(dfs, ignore_index=True)

# correct PIDs to HNLs
# IE, PID goes from +/-16 to +/- 19
merged_df['pid'] = merged_df['pid'].apply(lambda x: x*1.1875)

#Now add in the backgrounds:
merged_df = pd.concat([merged_df, background_df])

#Give the events unique indices:
merged_df.index.name = 'event_no'
# Save to a single Parquet file
merged_df.to_parquet("../parquet/truth/merged_file_truth.parquet", index=True)
```
